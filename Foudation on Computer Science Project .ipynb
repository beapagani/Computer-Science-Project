# 1 - Convert the App Sizes to a number
Dato che il termine 'Varies with device' potrebbe crare qualche misunderstanding, lo converto utilizzando un altro termine, in modo tale da poter tornare alla forma orinale

```python
GPS['Size']=[re.sub('Varies with device','999999999', size)for size in GPS['Size']]
GPS['Size']=[re.sub(',','.',size)for size in GPS['Size']]
```

```python
sizes_to_num_re=re.compile('(?P<amount>\d*\.*\d*)(?P<unit>\w*\+*)')

def unit_to_mult(unit):
    if unit == 'G':     #stay for Giga
        return 1000000000
    if unit == 'M':    #stay for Mega
        return 1000000
    if unit == 'k':    #stay for Kilo
          return 1000
    return 1
#consideraiamo queste unità di misura, nel caso venissero usate altre, decido fi ignorarle, dato che sono una quantità irrisoria

#la seguente funzione può comprendere il valore del formato dei dati usando Regex.
#se i dati non sono nel formato previsto restituiranno 'Not a Number'. Un esempio è 'Varies with device'
def to_numeric(elem): 
    m = sizes_to_num_re.search(elem)
    unit = m.group('unit')
    mult = unit_to_mult(unit)
    amount = float(m.group('amount'))
    return int(amount * mult)
```

```python
GPS['ExtSize'] = GPS['Size'].apply(to_numeric)
GPS.head()
```

Una volta che è stato convertito l' attributo Size, riporto Varies with device, in modo tale che venga preservato il suo significato

```python
GPS['Size'] = [re.sub('99999999', 'Varies with device', size) for size in GPS['Size']] #Reconverting the value '99999999' into 'Varies with device'
GPS['ExtSize'] = GPS['ExtSize'].apply(str)
GPS['ExtSize'] = [re.sub('99999999', 'Varies with device', size) for size in GPS['ExtSize']]
GPS[['App','Size','ExtSize']].head()
                             
```

```python
install_to_num = re.compile('(?P<number>[\d,]+)') #il formato è num,num...,num

#attraverso Regex si trova i numero nel formato
def installToNumber(installs):
    found = install_to_num.search(installs)
    if(found):
        replacedComma = found.group('number').replace(',','') #si sostituiscono le , non mettendo nessun valore
        return int(replacedComma)
    else:
        return np.nan
```

```python
GPS['InstallNumber']=GPS['Installs'].apply(installToNumber)
```

# 9 - For each genre, create a new column of the original dataframe.

The new columns must have boolean values (True if the app has a given genre)


Per risolvere questo task si è esplicitata una funzione ad hoc per assolvere questo compito.  Per i generi è stata utilizzata la colonna creata precedentemente, in modo tale che i generi vengano distinti 

```python
def create_dummies( df, colname ):
    col_dummies = pd.get_dummies(df[colname].apply(pd.Series).stack()).sum(level=0)
    col_dummies.drop(col_dummies.columns[0], axis=1, inplace=True)
    df = pd.concat([df, col_dummies.astype(bool)], axis=1)
    df.drop(colname, axis = 1, inplace = True )
    return df
```

```python
BoolGenres = create_dummies(GPS, 'GenresList')
BoolGenres[BoolGenres['App'] == 'Facebook'][['App','Genres', 'Art & Design', 'Pretend Play', 'Creativity']]
```

# 11 - For each app, compute the approximate income, obtain as a product of number of installs and price.

Si provvede come prima cosa a convertire il prezzo in numeri

```python
money_to_float = re.compile('\D*(?P<amount>\d+\.{0,1}\d*)')

def Convert(money):
    found = money_to_float.search(money)
    if found:
        return float(found.group('amount'))
    else:
        return None


GPS['PriceInNumber']=GPS.Price.apply(Convert)
```

Si crea una nuova colonna che contiene il numero approssimativo di Incomes

```python
GPS['approximate_income']=GPS.InstallNumber * GPS.PriceInNumber
```

```python
GPS[['App','Price','InstallNumber','PriceInNumber','approximate_income']][GPS.Price!='0'].head()
```
